```python
import pandas as pd
import numpy as np

#Number of interns
n = 40

#Generate UNIQUE Intern IDs
ids = set()
while len(ids) < n:
    ids.add("INT" + str(np.random.randint(10000, 99999)))
ids = list(ids)

data = []

for i in range(n):
    intern_id = ids[i]

    attendance = np.random.randint(50, 100)
    consistency = np.random.randint(40, 100)
    learning = np.random.randint(40, 100)

    tasks_assigned = 20
    tasks_completed = np.random.randint(5, 21)
    completion_rate = tasks_completed / tasks_assigned

    project_score = np.random.randint(40, 100)
    evaluation_score = np.random.randint(40, 100)

    login_freq = np.random.randint(1, 7)
    active_days = np.random.randint(5, 25)

    engagement = np.random.randint(1, 10)
    communication = np.random.randint(1, 10)
    discipline = np.random.randint(1, 10)

    #Performance Logic
    if completion_rate > 0.8 and engagement > 7 and evaluation_score > 75:
        performance = 2   #High
    elif completion_rate > 0.5 and evaluation_score > 60:
        performance = 1   #Medium
    else:
        performance = 0   #Low

    data.append([
        intern_id,
        attendance,
        consistency,
        learning,
        tasks_assigned,
        tasks_completed,
        completion_rate,
        project_score,
        evaluation_score,
        login_freq,
        active_days,
        engagement,
        communication,
        discipline,
        performance
    ])

#Column names
columns = [
    "intern_id",
    "attendance_percentage",
    "consistency_score",
    "learning_score",
    "tasks_assigned",
    "tasks_completed",
    "completion_rate",
    "project_score",
    "evaluation_score",
    "login_frequency",
    "active_days",
    "engagement_score",
    "communication_score",
    "discipline_score",
    "performance_label"
]

#Create DataFrame
df = pd.DataFrame(data, columns=columns)

#Show all rows
print("Total rows:", len(df))
df
```
